## Objetivo

Construir un notebook en Python donde se formule, implemente, simule e interprete un **modelo determinista sencillo**, siguiendo el ciclo de modelación visto en clase:

$$
\text{fenómeno real}
\longrightarrow
\text{variables y parámetros}
\longrightarrow
\text{supuestos}
\longrightarrow
\text{modelo matemático}
\longrightarrow
\text{simulación}
\longrightarrow
\text{comparación con datos}
\longrightarrow
\text{interpretación}
$$


## Instrucciones generales

Elaborar un notebook en Python que contenga el desarrollo completo de un modelo determinista. Puede elegir uno de los siguientes modelos:

1. **Modelo de crecimiento geométrico**
   $$P_{t+1}=P_t(1+r)$$

2. **Modelo de enfriamiento de Newton discreto**
   $$T_{t+1}=T_t-k(T_t-T_a)$$

3. **Modelo logístico discreto**
   $$P_{t+1}=P_t+rP_t\left(1-\frac{P_t}{K}\right)$$
   

4. **Modelo SIR básico discreto**
   $$
   S_{t+1}=S_t-\beta\frac{S_tI_t}{N}
   $$

   $$
   I_{t+1}=I_t+\beta\frac{S_tI_t}{N}-\gamma I_t
   $$
   $$
   R_{t+1}=R_t+\gamma I_t
   $$

También pueden proponer otro modelo determinista sencillo, siempre que justifiquen claramente sus variables, parámetros y supuestos.
